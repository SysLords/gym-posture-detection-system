```python
import pandas as pd
import numpy as np

angles= pd.read_csv('exercise_angles.csv')
```

```python
plank= angles[angles['exercise']=='plank']

# Select the necessary columns for plank detection
required_columns = [
    'right_ankle', 'left_ankle', 'right_hip', 'left_hip',
    'torso_bend', 'neck', 'right_shoulder', 'left_shoulder',
    'right_arm_torso', 'left_arm_torso'
]
plank_data = plank[required_columns]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
# Remove rows with NaN values

# Display the cleaned data
print(plank_data)

average_values = plank_data.mean(skipna=True)

# Display the results
print("Average Values:")
print(average_values)
```
